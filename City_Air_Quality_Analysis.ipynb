“The AQI sub-index for all necessary pollutants is calculated using CPCB-defined concentration breakpoints.
Linear interpolation was applied within each breakpoint range to convert Pollutants concentrations into AQI sub-indices.
This approach ensures continuity and reflects the gradual increase in health risk with rising particulate matter concentration.”

```python
def calculate_sub_index_pm25(pm25):
    if pm25 <= 30:
        return pm25 * 50 / 30
    elif pm25 <= 60:
        return ((pm25-30) * 50 / 30) + 50
    elif pm25 <= 90:
        return ((pm25-60) * 100 / 30) + 100
    elif pm25 <= 120:
        return ((pm25-90) * 100 / 30) + 200
    elif pm25 <= 250:
        return ((pm25-120) * 100 / 130) + 300
    else:
        return ((pm25-250) * 100 / 130) + 400
```

```python
def calculate_sub_index_pm10(pm10):
    if pm10 <= 50:
        return pm10
    elif pm10 <= 100:
        return pm10
    elif pm10 <= 250:
        return ((pm10-100)*100/150)+100
    elif pm10 <= 350:
        return ((pm10-250)*100/100)+200
    elif pm10 <= 430:
        return ((pm10-350)*100/80)+300
    else:
        return ((pm10-430)*100/80)+400
```

```python
def calculate_sub_index_no2(no2):
    if no2 <= 40:
        return no2 * 50 / 40
    elif no2 <= 80:
        return ((no2-40)*50/40)+50
    elif no2 <= 180:
        return ((no2-80)*100/100)+100
    elif no2 <= 280:
        return ((no2-180)*100/100)+200
    elif no2 <= 400:
        return ((no2-280)*100/120)+300
    else:
        return ((no2-400)*100/120)+400
```

```python
def calculate_sub_index_so2(so2):
    if so2 <= 40:
        return so2 * 50 / 40
    elif so2 <= 80:
        return ((so2 - 40) * 50 / 40) + 50
    elif so2 <= 380:
        return ((so2 - 80) * 100 / 300) + 100
    elif so2 <= 800:
        return ((so2 - 380) * 100 / 420) + 200
    elif so2 <= 1600:
        return ((so2 - 800) * 100 / 800) + 300
    else:
        return ((so2 - 1600) * 100 / 800) + 400
```

```python
def calculate_sub_index_co(co):
    if co <= 1.0:
        return co * 50 / 1.0
    elif co <= 2.0:
        return ((co - 1.0) * 50 / 1.0) + 50
    elif co <= 10:
        return ((co - 2.0) * 100 / 8.0) + 100
    elif co <= 17:
        return ((co - 10) * 100 / 7) + 200
    elif co <= 34:
        return ((co - 17) * 100 / 17) + 300
    else:
        return ((co - 34) * 100 / 17) + 400
```

```python
def calculate_sub_index_o3(o3):
    if o3 <= 50:
        return o3 * 50 / 50
    elif o3 <= 100:
        return ((o3 - 50) * 50 / 50) + 50
    elif o3 <= 168:
        return ((o3 - 100) * 100 / 68) + 100
    elif o3 <= 208:
        return ((o3 - 168) * 100 / 40) + 200
    elif o3 <= 748:
        return ((o3 - 208) * 100 / 540) + 300
    else:
        return ((o3 - 748) * 100 / 540) + 400
```

CALCULATE SUB-INDICES IN DATAFRAME

```python
df['PM2.5_Sub'] = df['PM2.5'].apply(calculate_sub_index_pm25)
df['PM10_Sub'] = df['PM10'].apply(calculate_sub_index_pm10)
df['NO2_Sub'] = df['NO2'].apply(calculate_sub_index_no2)
df['SO2_Sub'] = df['SO2'].apply(calculate_sub_index_so2)
df['CO_Sub']  = df['CO'].apply(calculate_sub_index_co)
df['O3_Sub']  = df['O3'].apply(calculate_sub_index_o3)
```

FINAL AQI CALCULATION

```python
df['Calculated_AQI'] = df[
    ['PM2.5_Sub','PM10_Sub','NO2_Sub','SO2_Sub','CO_Sub','O3_Sub']
].max(axis=1)
```

CREATE A COMPARISON DATAFRAME (ONLY WHERE AQI EXISTS)

```python
aqi_compare = df[df['AQI'].notna()][['City', 'Date', 'AQI', 'Calculated_AQI']]
aqi_compare.head()
```

BASIC DIFFERENCE ANALYSIS

```python
aqi_compare['AQI_Diff'] = aqi_compare['AQI'] - aqi_compare['Calculated_AQI']
aqi_compare['AQI_Diff'].describe()
```

CHECK — CORRELATION

```python
aqi_compare[['AQI', 'Calculated_AQI']].corr()
```

FILL MISSING AQI VALUES

```python
df['AQI'] = df['AQI'].fillna(df['Calculated_AQI'])
```

CREATE AQI BUCKETS (CPCB STANDARD)

```python
def aqi_bucket(aqi):
    if aqi <= 50:
        return 'Good'
    elif aqi <= 100:
        return 'Satisfactory'
    elif aqi <= 200:
        return 'Moderate'
    elif aqi <= 300:
        return 'Poor'
    elif aqi <= 400:
        return 'Very Poor'
    else:
        return 'Severe'

df['AQI_Bucket'] = df['AQI'].apply(aqi_bucket)
```

```python
#Fill AQI
df['Final_AQI'] = df['AQI'].fillna(df['Calculated_AQI'])
```

```python
#Create bucket from Final AQI
df['Final_AQI_Bucket'] = df['Final_AQI'].apply(aqi_bucket)
```

Drop helper columns (clean dataset)

```python
df.drop(columns=[
    'PM2.5_Sub','PM10_Sub','NO2_Sub','SO2_Sub','CO_Sub','O3_Sub','Calculated_AQI'
], inplace=True)
```

CITY-WISE YEARLY AVERAGE

“Ahmedabad emerges as the most severely polluted city with extreme AQI spikes, while other major Indian cities show gradual air quality improvement, especially after 2018.”
